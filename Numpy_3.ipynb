# Advance Operation with Business Examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],
    [2, 650000, 580000, 230000, 260000],
    [3, 250000, 180000, 270000, 220000],
    [4, 160000, 150000, 290000, 240000],
    [5, 170000, 190000, 260000, 270000]
])

print("=== Zomato sales analysis ===")
print("\n Sales data shape ", sales_data.shape)
print("\n Sample daata for 1st 3 restaurant: ", sales_data[0:3])
```

```python
# total sales per year
print(np.sum(sales_data, axis = 0))
yearly_total = np.sum(sales_data[:, 1:],axis = 1)
print(yearly_total)

# Minimum sales per restaurant
min_sales = np.min(sales_data[:,1:], axis = 0)
print(min_sales)

# Maximum sales per year
max_sales = np.max(sales_data[:,1:], axis = 0)
print(max_sales)

# Average sales per restaurant
avg_sales = np.mean(sales_data[:,1:], axis = 1)
print(avg_sales)
```

```python
cumsum = np.cumsum(sales_data[:,1:], axis = 1)
print(cumsum)

plt.figure(figsize = (5,6))
plt.plot(np.mean(cumsum, axis = 0))
plt.title("Average cumulative sales accross all restaurant ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector addition: ", vector1 + vector2)
print("\n Multiplication: ", vector1 * vector2)
print("\n Dot Product: ", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```

```python
restaurant_types = np.array(['biryani', 'chinese', 'pizza', 'burgur', 'cafe'])
vectorized_upper = np.vectorize(str.upper)
print("Vectorized Upper", vectorized_upper(restaurant_types))
```
